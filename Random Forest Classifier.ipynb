```python
Map.setCenter(98.54, 25.85, 11); 
//var mine = ee.Geometry.Point(98.54667827723028,25.854050360113273);
var mine = L7.filterBounds(ROI)
.filterDate("2004-01-01","2005-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

print(mine)

Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True color');

Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'False color');

// Extract the near infrared and red bands.
var nir = mine.select('B4');
var red = mine.select('B3');

// Calculate the numerator and the denominator using subtraction and addition respectively.
var numerator = nir.subtract(red);
var denominator = nir.add(red);

// Now calculate NDVI.
var ndvi2005 = numerator.divide(denominator);

// Add the layer to our map with a palette.
var vegPalette = ['red', 'white', 'green'];
Map.addLayer(ndvi2005, {
    min: -1,
    max: 1,
    palette: vegPalette
}, '2005 NDVI Manual');





// Combine training feature collections.

var trainingFeatures = ee.FeatureCollection([
    Mine, NoneMine
]).flatten();


// Define prediction bands.
var predictionBands = [
    'B1', 'B2', 'B3', 'B4', 'B5', 'B7',
    'B8'
];

// Sample training points.
var classifierTraining = mine.select(predictionBands)
    .sampleRegions({
    collection: trainingFeatures,
    properties: ['LC'],
    scale: 30
});

//////////////// CART Classifier ///////////////////

// Train a CART Classifier.
var classifier = ee.Classifier.smileCart().train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify the Landsat image.
var classified = mine.select(predictionBands).classify(classifier);

// Define classification image visualization parameters.
var classificationVis = {
    min: 0,
    max: 1,
    palette: ['red', 'green']
};

// Add the classified image to the map.
Map.addLayer(classified, classificationVis, 'CART classified');


/////////////// Random Forest Classifier /////////////////////

// Train RF classifier.
var RFclassifier = ee.Classifier.smileRandomForest(50).train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify Landsat image.
var RFclassified = mine.select(predictionBands).classify(
    RFclassifier);

// Add classified image to the map.
Map.addLayer(RFclassified, classificationVis, 'RF classified');

print(RFclassifier.confusionMatrix()); 

//print(classifier.confusionMatrix()); 



//Export GeoTiff

Export.image.toDrive({
 image: RFclassified,
 description: "Classified2005",
 folder: "GEE Classified",
 fileNamePrefix: "Classified2005_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});


Map.setCenter(98.54, 25.85, 11); 
//var mine = ee.Geometry.Point(98.54667827723028,25.854050360113273);
var mine = L7.filterBounds(ROI)
.filterDate("2009-01-01","2010-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

print(mine)

Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True color');

Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'False color');

// Extract the near infrared and red bands.
var nir = mine.select('B4');
var red = mine.select('B3');

// Calculate the numerator and the denominator using subtraction and addition respectively.
var numerator = nir.subtract(red);
var denominator = nir.add(red);

// Now calculate NDVI.
var ndvi2010 = numerator.divide(denominator);

// Add the layer to our map with a palette.
var vegPalette = ['red', 'white', 'green'];
Map.addLayer(ndvi2010, {
    min: -1,
    max: 1,
    palette: vegPalette
}, '2010 NDVI Manual');





// Combine training feature collections.

var trainingFeatures = ee.FeatureCollection([
    Mine, NoneMine
]).flatten();


// Define prediction bands.
var predictionBands = [
    'B1', 'B2', 'B3', 'B4', 'B5', 'B7',
    'B8'
];

// Sample training points.
var classifierTraining = mine.select(predictionBands)
    .sampleRegions({
    collection: trainingFeatures,
    properties: ['LC'],
    scale: 30
});

//////////////// CART Classifier ///////////////////

// Train a CART Classifier.
var classifier = ee.Classifier.smileCart().train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify the Landsat image.
var classified = mine.select(predictionBands).classify(classifier);

// Define classification image visualization parameters.
var classificationVis = {
    min: 0,
    max: 1,
    palette: ['red', 'green']
};

// Add the classified image to the map.
Map.addLayer(classified, classificationVis, 'CART classified');


/////////////// Random Forest Classifier /////////////////////

// Train RF classifier.
var RFclassifier = ee.Classifier.smileRandomForest(50).train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify Landsat image.
var RFclassified = mine.select(predictionBands).classify(
    RFclassifier);

// Add classified image to the map.
Map.addLayer(RFclassified, classificationVis, 'RF classified');

print(RFclassifier.confusionMatrix()); 

//print(classifier.confusionMatrix()); 



//Export GeoTiff

Export.image.toDrive({
 image: RFclassified,
 description: "Classified2010",
 folder: "GEE Classified",
 fileNamePrefix: "Classified2010_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});


Map.setCenter(98.54, 25.85, 11); 
//var mine = ee.Geometry.Point(98.54667827723028,25.854050360113273);
var mine = L8.filterBounds(ROI)
.filterDate("2014-01-01","2015-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

print(mine)

Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'True color');

Map.addLayer(mine, {bands:["B5","B4","B3"]}, 'False color');

// Extract the near infrared and red bands.
var nir = mine.select('B5');
var red = mine.select('B4');

// Calculate the numerator and the denominator using subtraction and addition respectively.
var numerator = nir.subtract(red);
var denominator = nir.add(red);

// Now calculate NDVI.
var ndvi2015 = numerator.divide(denominator);

// Add the layer to our map with a palette.
var vegPalette = ['red', 'white', 'green'];
Map.addLayer(ndvi2015, {
    min: -1,
    max: 1,
    palette: vegPalette
}, '2015 NDVI Manual');



// Use normalizedDifference to calculate NDWI
var ndwi2015 = mine.normalizedDifference(['B8', 'B11']);
var waterPalette = ['white', 'blue'];
Map.addLayer(ndwi2015, {
    min: -0.5,
    max: 1,
    palette: waterPalette
}, '2015 NDWI');


// Combine training feature collections.

var trainingFeatures = ee.FeatureCollection([
    Mine, NoneMine
]).flatten();


// Define prediction bands.
var predictionBands = [
    'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7',
    'B10', 'B11'
];

// Sample training points.
var classifierTraining = mine.select(predictionBands)
    .sampleRegions({
    collection: trainingFeatures,
    properties: ['LC'],
    scale: 30
});

//////////////// CART Classifier ///////////////////

// Train a CART Classifier.
var classifier = ee.Classifier.smileCart().train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify the Landsat image.
var classified = mine.select(predictionBands).classify(classifier);

// Define classification image visualization parameters.
var classificationVis = {
    min: 0,
    max: 1,
    palette: ['red', 'green']
};

// Add the classified image to the map.
Map.addLayer(classified, classificationVis, 'CART classified');


/////////////// Random Forest Classifier /////////////////////

// Train RF classifier.
var RFclassifier = ee.Classifier.smileRandomForest(50).train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify Landsat image.
var RFclassified = mine.select(predictionBands).classify(
    RFclassifier);

// Add classified image to the map.
Map.addLayer(RFclassified, classificationVis, 'RF classified');

print(RFclassifier.confusionMatrix()); 

//print(classifier.confusionMatrix()); 

//Export GeoTiff

Export.image.toDrive({
 image: RFclassified,
 description: "Classified2015",
 folder: "GEE Classified",
 fileNamePrefix: "Classified2015_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});

Map.setCenter(98.54, 25.85, 11); 
//var mine = ee.Geometry.Point(98.54667827723028,25.854050360113273);
var mine = L8.filterBounds(ROI)
.filterDate("2019-01-01","2020-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

print(mine)

Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'True color');

Map.addLayer(mine, {bands:["B5","B4","B3"]}, 'False color');

// Extract the near infrared and red bands.
var nir = mine.select('B5');
var red = mine.select('B4');

// Calculate the numerator and the denominator using subtraction and addition respectively.
var numerator = nir.subtract(red);
var denominator = nir.add(red);

// Now calculate NDVI.
var ndvi2020 = numerator.divide(denominator);

// Add the layer to our map with a palette.
var vegPalette = ['red', 'white', 'green'];
Map.addLayer(ndvi2020, {
    min: -1,
    max: 1,
    palette: vegPalette
}, '2020 NDVI Manual');



// Use normalizedDifference to calculate NDWI
var ndwi2020 = mine.normalizedDifference(['B8', 'B11']);
var waterPalette = ['white', 'blue'];
Map.addLayer(ndwi2020, {
    min: -0.5,
    max: 1,
    palette: waterPalette
}, '2020 NDWI');


// Combine training feature collections.

var trainingFeatures = ee.FeatureCollection([
    Mine, NoneMine
]).flatten();


// Define prediction bands.
var predictionBands = [
    'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7',
    'B10', 'B11'
];

// Sample training points.
var classifierTraining = mine.select(predictionBands)
    .sampleRegions({
    collection: trainingFeatures,
    properties: ['LC'],
    scale: 30
});

//////////////// CART Classifier ///////////////////

// Train a CART Classifier.
var classifier = ee.Classifier.smileCart().train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify the Landsat image.
var classified = mine.select(predictionBands).classify(classifier);

// Define classification image visualization parameters.
var classificationVis = {
    min: 0,
    max: 1,
    palette: ['red', 'green']
};

// Add the classified image to the map.
Map.addLayer(classified, classificationVis, 'CART classified');


/////////////// Random Forest Classifier /////////////////////

// Train RF classifier.
var RFclassifier = ee.Classifier.smileRandomForest(50).train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify Landsat image.
var RFclassified = mine.select(predictionBands).classify(
    RFclassifier);

// Add classified image to the map.
Map.addLayer(RFclassified, classificationVis, 'RF classified');

print(RFclassifier.confusionMatrix()); 

//print(classifier.confusionMatrix()); 



//Export GeoTiff

Export.image.toDrive({
 image: RFclassified,
 description: "Classified2020",
 folder: "GEE Classified",
 fileNamePrefix: "Classified2020_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});



```
