```python
//************2020 Analysis Landsat 8*********************************
//Location for bound
//Loading Landasat 8 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L8.filterBounds(ROI)
.filterDate("2019-01-01","2020-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'True Color 2020');
Map.addLayer(mine, {bands:["B5","B4","B3"]}, 'FCC 2020');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B4_RED = mine.select('B4');
var B5_NIR = mine.select('B5');

//GEE in-built NDVI Calculation
var ndvi1 = B5_NIR.subtract(B4_RED).divide(B5_NIR.add(B4_RED));

//Dispalying NDVI in GEE
var ndvi1_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi1, {min: -0.1, max: 1.0, palette: ndvi1_palette}, 'NDVI 2020');

//Calculating SAVI
var savi = mine.expression(
    '1.5 * ((NIR - RED) / (NIR + RED + 0.5))', {
      'NIR': mine.select('B5'),
      'RED': mine.select('B4')
  }
);

//Dispalying SAVI
var savi_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(savi, {min: -0.1, max: 1.0, palette: savi_palette}, 'SAVI 2020');


// compute MSAVI using expression
var msavi = mine.expression(
  '(2 * NIR + 1 - sqrt(pow((2 * NIR + 1), 2) - 8 * (NIR - RED)) ) / 2', 
  {
    'NIR': mine.select('B5'), 
    'RED': mine.select('B4')
  }
);

//Dispalying MSAVI in GEE
var msavi_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(msavi, {min: -0.1, max: 1.0, palette: msavi_palette}, 'MSAVI 2020 (expression)');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi1,
 description: "NDVI2020Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2020Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});


//************2019 Analysis Landsat 8*********************************
//Location for bound
//Loading Landasat 8 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L8.filterBounds(ROI)
.filterDate("2018-01-01","2019-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'True Color 2019');
Map.addLayer(mine, {bands:["B5","B4","B3"]}, 'FCC 2019');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B4_RED = mine.select('B4');
var B5_NIR = mine.select('B5');

//GEE in-built NDVI Calculation
var ndvi2 = B5_NIR.subtract(B4_RED).divide(B5_NIR.add(B4_RED));

//Dispalying NDVI in GEE
var ndvi2_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi2, {min: -0.1, max: 1.0, palette: ndvi2_palette}, 'NDVI 2019');


//Export GeoTiff

Export.image.toDrive({
 image: ndvi2,
 description: "NDVI2019Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2019Myanmar_2",
 scale: 30,
 fileFormat: "GeoTiff"
});


//************2017 Analysis Landsat 8*********************************
//Location for bound
//Loading Landasat 8 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L8.filterBounds(ROI)
.filterDate("2016-01-01","2017-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'True Color 2017');
Map.addLayer(mine, {bands:["B5","B4","B3"]}, 'FCC 2017');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B4_RED = mine.select('B4');
var B5_NIR = mine.select('B5');

//GEE in-built NDVI Calculation
var ndvi3 = B5_NIR.subtract(B4_RED).divide(B5_NIR.add(B4_RED));

//Dispalying NDVI in GEE
var ndvi3_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi3, {min: -0.1, max: 1.0, palette: ndvi3_palette}, 'NDVI 2017');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi3,
 description: "NDVI2017Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2017Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});


//************2015 Analysis Landsat 8*********************************
//Location for bound
//Loading Landasat 8 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L8.filterBounds(ROI)
.filterDate("2014-01-01","2015-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'True Color 2015');
Map.addLayer(mine, {bands:["B5","B4","B3"]}, 'FCC 2015');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B4_RED = mine.select('B4');
var B5_NIR = mine.select('B5');

//GEE in-built NDVI Calculation
var ndvi4 = B5_NIR.subtract(B4_RED).divide(B5_NIR.add(B4_RED));

//Dispalying NDVI in GEE
var ndvi4_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi4, {min: -0.1, max: 1.0, palette: ndvi4_palette}, 'NDVI 2015');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi4,
 description: "NDVI2015Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2015Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});

//************2013 Analysis Landsat 7*********************************
//Location for bound
//Loading Landasat 7 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L7.filterBounds(ROI)
.filterDate("2012-01-01","2013-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True Color 2013');
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'FCC 2013');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B3_RED = mine.select('B3');
var B4_NIR = mine.select('B4');

//GEE in-built NDVI Calculation
var ndvi5 = B4_NIR.subtract(B3_RED).divide(B4_NIR.add(B3_RED));

//Dispalying NDVI in GEE
var ndvi5_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi5, {min: -0.1, max: 1.0, palette: ndvi5_palette}, 'NDVI 2013');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi5,
 description: "NDVI2013Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2013Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});

//************2010 Analysis Landsat 7*********************************
//Location for bound
//Loading Landasat 7 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L7.filterBounds(ROI)
.filterDate("2009-01-01","2010-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True Color 2010');
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'FCC 2010');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B3_RED = mine.select('B3');
var B4_NIR = mine.select('B4');

//GEE in-built NDVI Calculation
var ndvi6 = B4_NIR.subtract(B3_RED).divide(B4_NIR.add(B3_RED));

//Dispalying NDVI in GEE
var ndvi6_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi6, {min: -0.1, max: 1.0, palette: ndvi5_palette}, 'NDVI 2010');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi6,
 description: "NDVI2010Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2010Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});

//************2005 Analysis Landsat 5*********************************
//Location for bound
//Loading Landasat 8 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L7.filterBounds(ROI)
.filterDate("2004-01-01","2005-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True Color 2005');
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'FCC 2005');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B3_RED = mine.select('B3');
var B4_NIR = mine.select('B4');

//GEE in-built NDVI Calculation
var ndvi7 = B4_NIR.subtract(B3_RED).divide(B4_NIR.add(B3_RED));

//Dispalying NDVI in GEE
var ndvi7_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi7, {min: -0.1, max: 1.0, palette: ndvi7_palette}, 'NDVI 2005');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi7,
 description: "NDVI2005Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2005Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});


//************2002 Analysis Landsat 5*********************************
//Location for bound
//Loading Landasat 8 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L7.filterBounds(ROI)
.filterDate("2001-01-01","2002-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True Color 2002');
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'FCC 2002');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B3_RED = mine.select('B3');
var B4_NIR = mine.select('B4');

//GEE in-built NDVI Calculation
var ndvi8 = B4_NIR.subtract(B3_RED).divide(B4_NIR.add(B3_RED));

//Dispalying NDVI in GEE
var ndvi8_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi8, {min: -0.1, max: 1.0, palette: ndvi8_palette}, 'NDVI 2002');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi8,
 description: "NDVI2002Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2002Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});


//************2000 Analysis Landsat 5*********************************
//Location for bound
//Loading Landasat 8 data
//Dates of interest
//Filtering cloud cover

Map.setCenter(98.54, 25.85, 11); 
var mine = L7.filterBounds(ROI)
.filterDate("2000-01-01","2002-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

//Displaying True colour
Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True Color 2000');
Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'FCC 2000');

//*************NDVI Calculation*****************//

//Creatimg band variables
var B3_RED = mine.select('B3');
var B4_NIR = mine.select('B4');

//GEE in-built NDVI Calculation
var ndvi9 = B4_NIR.subtract(B3_RED).divide(B4_NIR.add(B3_RED));

//Dispalying NDVI in GEE
var ndvi9_palette = 'FFFFFF, CE7E45, DF923D, FCD163, 998718, 74A901, 66A000, 529400,' +
'3E8601, 207401, 056201, 004C00, 023C00, 023801, 012E01, 011301';

Map.addLayer(ndvi9, {min: -0.1, max: 1.0, palette: ndvi9_palette}, 'NDVI 2000');

//Export GeoTiff

Export.image.toDrive({
 image: ndvi9,
 description: "NDVI2000Myanmar",
 folder: "GEE Export2",
 fileNamePrefix: "NDVI2000Myanmar_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});


# Creating a Time-Series Chart

var startyear = 2000;
var endyear = 2020;

var years = ee.List.sequence(startyear, endyear);

var L7 = ee.ImageCollection("LANDSAT/LE7_L1T_ANNUAL_NDVI")
        .filterDate("2000-01-01", "2013-12-31")
        .select("NDVI")
        
        
var L8 = ee.ImageCollection("LANDSAT/LC8_L1T_ANNUAL_NDVI")
        .filterDate("2014-01-01", "2020-12-31")
        .select("NDVI")
        

var dataset = L7.merge(L8)

var annual_collection = ee.ImageCollection.fromImages(years.map(function (y){
    var annual = dataset.filter(ee.Filter.calendarRange(y, y, 'year'))
                      .mean().clip(ROI);
    return annual.set('year',y)
    }));
    
print(annual_collection)

var meanNDVI = dataset.reduce(ee.Reducer.mean())

var final = meanNDVI.clip(ROI)

var colorizedVis = {
  min: 0.0,
  max: 1.0,
  palette: ['FFFFFF', 'CE7E45', 'DF923D', 'FCD163', '998718', '74A901', 
            '66A000', '529400', '3E8601', '207401', '056201', '004C00', 
            '023C00', '023801', '012E01', '011301'                      
           ],  
         };
         
var chart = ui.Chart.image.series({
  imageCollection:dataset,
  region:ROI,
  reducer: ee.Reducer.mean(),
  scale:30,
   xProperty: 'system:time_start'}).setOptions({
    title: "NDVI Yearly Chart of Mynamar Mining Site from year 2000 to 2020", 
    vAxis: {title: 'NDVI Value'},
    hAxis: {title: 'Year'},
   trendlines: {0: {color: '#aa3319', //type: 'exponetial',
    visibleInLengend: true}},
    series: {0: {color: '#161616'}
      
    }
  })

print(chart)
         
Map.addLayer(final, colorizedVis, 'Myanamar NDVI');
```

```python
Map.setCenter(98.54, 25.85, 11); 
//var mine = ee.Geometry.Point(98.54667827723028,25.854050360113273);
var mine = L7.filterBounds(ROI)
.filterDate("2004-01-01","2005-12-31")
.filterMetadata("CLOUD_COVER", "less_than", 1)
.mean()
.clip(ROI); 

print(mine)

Map.addLayer(mine, {bands:["B3","B2","B1"]}, 'True color');

Map.addLayer(mine, {bands:["B4","B3","B2"]}, 'False color');

// Extract the near infrared and red bands.
var nir = mine.select('B4');
var red = mine.select('B3');

// Calculate the numerator and the denominator using subtraction and addition respectively.
var numerator = nir.subtract(red);
var denominator = nir.add(red);

// Now calculate NDVI.
var ndvi2005 = numerator.divide(denominator);

// Add the layer to our map with a palette.
var vegPalette = ['red', 'white', 'green'];
Map.addLayer(ndvi2005, {
    min: -1,
    max: 1,
    palette: vegPalette
}, '2005 NDVI Manual');





// Combine training feature collections.

var trainingFeatures = ee.FeatureCollection([
    Mine, NoneMine
]).flatten();


// Define prediction bands.
var predictionBands = [
    'B1', 'B2', 'B3', 'B4', 'B5', 'B7',
    'B8'
];

// Sample training points.
var classifierTraining = mine.select(predictionBands)
    .sampleRegions({
    collection: trainingFeatures,
    properties: ['LC'],
    scale: 30
});

//////////////// CART Classifier ///////////////////

// Train a CART Classifier.
var classifier = ee.Classifier.smileCart().train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify the Landsat image.
var classified = mine.select(predictionBands).classify(classifier);

// Define classification image visualization parameters.
var classificationVis = {
    min: 0,
    max: 1,
    palette: ['red', 'green']
};

// Add the classified image to the map.
Map.addLayer(classified, classificationVis, 'CART classified');


/////////////// Random Forest Classifier /////////////////////

// Train RF classifier.
var RFclassifier = ee.Classifier.smileRandomForest(50).train({
    features: classifierTraining,
    classProperty: 'LC',
    inputProperties: predictionBands
});

// Classify Landsat image.
var RFclassified = mine.select(predictionBands).classify(
    RFclassifier);

// Add classified image to the map.
Map.addLayer(RFclassified, classificationVis, 'RF classified');

print(RFclassifier.confusionMatrix()); 

//print(classifier.confusionMatrix()); 



//Export GeoTiff

Export.image.toDrive({
 image: RFclassified,
 description: "Classified2005",
 folder: "GEE Classified",
 fileNamePrefix: "Classified2005_2",
 scale: 30,
 shardSize: 100,
 fileDimensions: 5000,
 fileFormat: "GeoTiff"
});
```
